#1. ) Determine what clubs have the highest average ranking. 
#2.) Out of those clubs, what nationalities are most common?
#The dataset being used is the Fifa 19 dataset from Kaggle and the analytical tools being used are pandas, matplotlib, and python. 

###Dowload datset from my Github and then upload it from your local drive. (fifa_19_data.csv)

###Import Pandas for data analysis

```
import pandas as pd
```

###After uploading file into Google Colab, import it into a dataframe

```
file = 'fifa_19_data.csv'
fifa_df = pd.read_csv(file)
```

###Create a copy of your dataframe

```
fifa = fifa_df
fifa.head()
```

```
fifa.shape, type(fifa) #18,207 rows & 89 columns, DataFrame Type
```

###Extract the clubs with the highest average overall ranking

```
teams = fifa.groupby('Club')['Overall'].mean().sort_values(ascending=False).head(11)
teams
```

```
teams.shape, type(teams)
```

###Create a list of the club names to add to your upcoming graph

```
clubs = ['Juventus', 'Napoli', 'Inter Milan', 'Real Madrid', 'AC Milan', 'Barcelona', 'PSG', 'Roma', 'Man. Utd.', 'Benfica', 'Bayern Munich']
clubs
```

###Import your data visualization tool & finally create your graph

```
import matplotlib.pyplot as plt
```

```
import matplotlib.patches as mpatches

fig = plt.figure(figsize=(14,8))
fig.patch.set_facecolor('#e6e6e6')
plt.style.use('classic')
for x,y in zip(teams, clubs):

    label = "{:.2f}".format(x)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(20,0), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or cent
plt.barh(clubs, teams, color=['#1aff1a', '#1aff1a', '#1aff1a', '#ffff1a', '#1aff1a', '#ffff1a', '#1a75ff', '#1aff1a', '#ff751a', '#ff1a1a', '#ff1ac6'], edgecolor='black')
#red-#ff1a1a, blue-#1a75ff, green-#1aff1a, yellow-'#ffff1a', purple-#ff1ac6, orange-#ff751a
plt.title('Fifa 19\'s Top 11 Clubs with the Highest Average Ranking', fontsize=19, loc='right')
plt.xlabel('Average Overall Ranking')
italy = mpatches.Patch(color='#1aff1a', label='Italy')
spain = mpatches.Patch(color='#ffff1a', label='Spain')
france = mpatches.Patch(color='#1a75ff', label='France')
england = mpatches.Patch(color='#ff751a', label='England')
portugal = mpatches.Patch(color='#ff1a1a', label='Portugal')
germany = mpatches.Patch(color='#ff1ac6', label='Germany')
plt.legend(handles=[italy, germany, france, england, spain, portugal], bbox_to_anchor=(0., 1.02, .3, .1), loc='lower left',
           ncol=2, mode='expand',borderaxespad=0, title='Countries for each Team')
#plt.savefig('Top_Ranked_Clubs_4')
plt.show()
```

```
teams
```

###Extract all the teams from above

```
fc = fifa['Club']
```

```
countries = fifa.loc[(fc == 'SL Benfica') | (fc == 'Roma') | (fc == 'Manchester United') | (fc == 'FC Bayern München') | (fc == 'Paris Saint-Germain') | (fc == 'FC Barcelona') | (fc == 'Junventus') | (fc == 'Real Madrid') | (fc == 'Napoli') | (fc == 'Milan') |(fc == 'Inter')]
countries
```

###Find the count of the various nationalities of the top clubs

```
top_nations = countries.groupby('Nationality')['Nationality'].value_counts().sort_values(ascending=False).head(13)
top_nations
```

###Create the labels for your pie plot

```
labels = ['Spain', 'Italy', 'France', 'Germany', 'Brazil', 'Argentina', 'Portugal', 'England', 'Croatia', 'Belgium', 'Netherlands', 'Uruguay', 'Serbia']
labels
```

###Plot your second graph

```
#red-#ff1a1a, blue-#1a75ff, green-#1aff1a, yellow-'#ffff1a', purple-#ff1ac6, orange-#ff751a

plt.figure(figsize=(10,10))
plt.pie(top_nations, labels=labels, autopct=lambda p: '{:.0f}'.format(p), shadow=True, colors=['#ffff1a', '#1a75ff', '#ff1ac6', '#1aff1a', '#ff751a', '#c61aff', '#ff1a1a'], textprops={'fontsize': 18})
plt.title('Most Common Nationalities in Fifa 19\'s Top 11 Clubs', fontsize=20)
plt.savefig('pie_plot_fifa')
plt.show()
```
